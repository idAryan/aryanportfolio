```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=Image.open('prof_pic7.png').convert('L')
```

```python
a=np.array(im)
```

```python
a
```

```python
threshold=128
binary_image=(a>threshold).astype(int)
```

```python
plt.imshow(binary_image, cmap='gray')
plt.title("Aryan Sharma")
plt.axis('off')
plt.show()
```

```python
im=Image.open('prof_pic1.png').convert('L')
small_img=im.resize((128,128))
a_small=np.array(small_img)
binary_image=(a_small>threshold).astype(int)
plt.figure(figsize=(6,6))
plt.imshow(binary_image,cmap='gray')
plt.axis('off')
rows,cols=binary_image.shape
```

```python
plt.figure(figsize=(6,6))
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
rows,cols=binary_image.shape
for i in range(rows):
    for j in range(cols):
        plt.text(j,i,str(binary_image[i,j]), ha='center', va='center', color='red',fontsize=6)
plt.title('Aryan Sharma')
plt.show()
```
